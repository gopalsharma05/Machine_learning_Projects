```python
r=r.drop(columns=['Outcome'])
r_col=r.columns
print(r.shape,r.columns)
```

```python
# As maximum of different columns have large difference . So we will perform nomalization 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data1=sc.fit_transform(data)
r=sc.fit_transform(r)
```

```python
data1=pd.DataFrame(data1,columns =data.columns)  # converting numpy to pandas dataframe with columns names

# looking at heatmap correlation matrix for feature selection

# we can also use PCA and plot importance of XGBoost for feature selecetion

import seaborn as sns
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data1.corr(), annot=True, linewidths=.5, fmt= '.1f')
```

```python
 #by loooking at heatmap we can see that perimeter_mean,radius_mean and area_mean are strongly  correlated
# i am keeping area_mean and removing other two. Also we can see that std_dev of radius ,perimeter,area are strongly correlated
# so i am kepping area.Also we can see that worst of radius ,perimeter,area are correlated so i am kepping area . Also
# worst_area ,area_mean are correlated , i keeping area mean. from worst_texture and texture_mean i am keeping texture mean
# we can see concave_point_mean and concavity mean is correlated. I am keeping cocavity_mean. Also Worst_fractal_dimension AND 
# fractal_dimension_mean  are correlated , i am keeping fractal_dimension_mean .


#I AM CONSIDERING ONLY STRONGLY CORRELATED FEATURE i.e VALUE=1 and VALUE=0.9


data1=data1.drop(columns=['perimeter_mean','radius_mean','radius_std_dev','perimeter_std_dev','Worst_perimeter','Worst_radius','Worst_area'
                        ,'Worst_texture','concave_points_mean','Worst_fractal_dimension'])
data1.shape
```

```python
#splitting in test and train datasets
from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_test1 = train_test_split(data1, y, test_size=0.2, random_state=42)
```

```python
# as we have very less data...so i will use simple model else other model(which req. many parameter can cause overfitting)
# so i will start with linear regression

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train1, y_train1)

# Predicting the Test set results
y_pred = regressor.predict(x_test1)

from sklearn import metrics
print(metrics.mean_absolute_error(y_test1, y_pred))
print(metrics.mean_squared_error(y_test1, y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test1, y_pred)))
 

```

```python
# Fitting Decision Tree Regression to the dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x_train1, y_train1)

# Predicting on test dataset
y_pred = regressor.predict(x_test1)


from sklearn import metrics
print(metrics.mean_absolute_error(y_test1, y_pred))
print(metrics.mean_squared_error(y_test1, y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test1, y_pred)))
```

```python
# Fitting Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 5, random_state = 0)
regressor.fit(x_train1, y_train1)                 #choose more no of n_estimator for better accuration i.e 
                                    #at 300 predicted value of 6.5 is ~ 160000

# Predicting test data
y_pred = regressor.predict(x_test1)

from sklearn import metrics
print(metrics.mean_absolute_error(y_test1, y_pred))
print(metrics.mean_squared_error(y_test1, y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test1, y_pred)))
```

```python
#DUE TO VERY LESS DATA PCA not give better result with xgboost,but better than above models i.e With PCA XGboost give 18% error

# #applying PCA for removing features further
# from sklearn.decomposition import PCA
# pca=PCA(n_components=10)    # with 10 features only
# x_train_pca=pca.fit_transform(x_train1)
# x_test_pca=pca.fit_transform(x_test1)

# using xgboost
import xgboost as xgb   

model = xgb.XGBRegressor(max_depth=1,learning_rate=0.01,reg_alpha=0.9,reg_lambda=1)
model.fit(x_train1,y_train1)

y_pred = model.predict(data=x_test1)

from sklearn import metrics
print(metrics.mean_absolute_error(y_test1, y_pred))
print(metrics.mean_squared_error(y_test1, y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test1, y_pred)))
```

```python
# converting in excel file  
r=x_test1
r["Outcome predicted"]=y_pred
r["Outcome actual"]=y_test1
r.to_excel("regression_submission.xlsx")
y_test1
```
