```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#taking excel file of data into data variable
data=pd.read_excel(r'D:\ML datasets\inovaccer intern\inno_ex.xlsx')
```

```python
print(data.shape,data.columns)
data
```

```python
data.info()   #no null values
```

```python
#removing colums Outcome and ID
y=data['Outcome']
data=data.drop(columns=['Outcome','ID'])
print(data.shape,y.shape)
```

```python
# column 'Lymph_Node_Status' is of object type and contain integer and some '?'. So we can replace '?' with 
# '0' (as most of value are 0) and converting columns to float
data
```

```python
for i in range(len(data['Lymph_Node_Status'])):
    if(data['Lymph_Node_Status'][i]=='?'):
        data['Lymph_Node_Status'][i]='0'

data['Lymph_Node_Status']=data['Lymph_Node_Status'].astype('int64')
```

```python
# we can see that there is almost 23% outcome are 'R' and rest are 'N'. So it is like unbalanced data. We can use 'SMOTE' 
# to get replicate data for 'R' outcome


import seaborn as sns
sns.countplot(y,label="Count") 
```

```python
# we can apply SMOTE for unbalanced data to make it balance so that good model we can train
y_smote=y
x_smote=data


from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state = 2) 
x_train_smote, y_train_smote = sm.fit_sample(x_smote, y_smote )
```

```python
sns.countplot(y_train_smote,label="Count")   # now we have equal 'R' and 'N' labels so it will train better now
```

```python
#size of training data set after applying smote algorithm
print(x_train_smote.shape,y_train_smote.shape)   #
```

```python
x_train_smote=pd.DataFrame(x_train_smote,columns = data.columns) # different colums have large difference betwwen max values so we have to normalize the columns
y_train_smote=pd.DataFrame(y_train_smote,columns = ['Outcome'])
 
```

```python
 print(x_train_smote.shape,y_train_smote.shape)
```

```python
# As maimum of different columns have large difference . So we will perform nomalization 
x_train_smote.describe()

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_smote=sc.fit_transform(x_train_smote)
```

```python
x_train_smote=pd.DataFrame(x_train_smote,columns =data.columns)  # converting numpy to pandas dataframe with columns names

# looking at heatmap correlation matrix for feature selection

# we can also use PCA and plot importance of XGBoost for feature selecetion

f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(x_train_smote.corr(), annot=True, linewidths=.5, fmt= '.1f')
```

```python
 #by loooking at heatmap we can see that perimeter_mean,radius_mean and area_mean are strongly  correlated
# i am keeping area_mean and removing other two. Also we can see that std_dev of radius ,perimeter,area are strongly correlated
# so i am kepping area.Also we can see that worst of radius ,perimeter,area are correlated so i am kepping area . Also
# worst_area ,area_mean are correlated , i keeping area mean. from worst_texture and texture_mean i am keeping texture mean

#I AM CONSIDERING ONLY STRONGLY CORRELATED FEATURE i.e VALUE=1 and VALUE=0.9


x_train_smote=x_train_smote.drop(columns=['perimeter_mean','radius_mean','radius_std_dev','perimeter_std_dev','Worst_perimeter','Worst_radius','Worst_area'
                        ,'Worst_texture'])
x_train_smote.shape
```

```python
# AGAIN plotting heat_map to cross check
f,ax = plt.subplots(figsize=(15, 15))
sns.heatmap(x_train_smote.corr(), annot=True, linewidths=.5, fmt= '.1f')



```

```python

# we can see concave_point_mean and concavity mean is correlated. I am keeping cocavity_mean
x_train_smote=x_train_smote.drop(columns=['concave_points_mean'])

x_train_smote.shape
```

```python
#splitting in test and train datasets
from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_test1 = train_test_split(x_train_smote, y_train_smote, test_size=0.25, random_state=42)
```

```python
# now training the models

#1. using xgboost

from xgboost import XGBClassifier
from xgboost import plot_importance
import xgboost as xgb


xgboost = XGBClassifier(max_depth=20, learning_rate=0.01, n_estimators=40, gamma=0, 
                        min_child_weight=1, subsample=0.6, colsample_bytree=0.6, reg_alpha=0.05)

xgboost.fit(x_train1, y_train1)


plot_importance(xgboost)    # we can also use this to reduce features
plt.show()


```

```python
#removing some more feature using feature importance from xgboost
x_train_xg=x_train1.drop(columns=['Worst_concavity','symmetry_std_dev','fractal_dimension_mean','Worst_compactness',
                                   'compactness_std_dev'])

x_test_xg=x_test1.drop(columns=['Worst_concavity','symmetry_std_dev','fractal_dimension_mean','Worst_compactness',
                                   'compactness_std_dev'])
```

```python
xgboost = XGBClassifier(max_depth=20, learning_rate=0.01, n_estimators=100,
                        min_child_weight=1, subsample=0.8, colsample_bytree=0.5, reg_alpha=0.05)

xgboost.fit(x_train_xg, y_train1)
y_pred = xgboost.predict(x_test_xg)


# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test1, y_pred)

print(cm,classification_report(y_test1, y_pred))    # recall is greater than precisoin , so less false negative
print(y_pred,y_test1)    # we can see that ....the XGBoost model after applyying SMOTE gives good result


#THIS MODEL GIVES HIGHEST ACCURACY
```

```python
print("accuracy from XGBoost is --> ",(67/76)*100)  
```

```python
#2. using KNN model

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 2, metric = 'minkowski', p = 1)
classifier.fit(x_train_xg, y_train1)    #with 19 features

# Predicting the Test set results
y_pred = classifier.predict(x_test_xg)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test1, y_pred)

print(cm,classification_report(y_test1, y_pred))    # recall is less than precisoin , so more false negative
```

```python
print("accuracy from KNN is --> ",(67/76)*100)  
```

```python
 

# 2. Now using Random forest

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 30, criterion = 'entropy', random_state = 0) # out of 15 best n_estimator is 6
classifier.fit(x_train_xg, y_train1) #with 19 features

# Predicting the Test set results
y_pred = classifier.predict(x_test_xg)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test1, y_pred)

print(cm,classification_report(y_test1, y_pred))

```

```python
print("accuracy from random forest is --> ",(66/76)*100)  
```

```python
 

# 1. using DECISION TREE

from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_train_xg, y_train1)    #with 19 features

y_pred = classifier.predict(x_test_xg)

from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test1, y_pred)


print(cm,classification_report(y_test1, y_pred))
```
